```python
import numpy as np
import matplotlib.pyplot as plt


# 加载数据
node_values = np.load('/home/admin/research/traffic/METR-LA/node_values.npy')
adj_mat = np.load('/home/admin/research/traffic/METR-LA/adj_mat.npy')

# 绘制时间序列图
plt.figure(figsize=(15, 5))
plt.plot(node_values[:1000, 0, 0], label='Node 0 Speed')  # 展示第一个节点的速度数据
plt.plot(node_values[:1000, 1, 0], label='Node 1 Speed')  # 展示第二个节点的速度数据
plt.title('Traffic Speed Time Series')
plt.xlabel('Time Steps')
plt.ylabel('Speed')
plt.legend()
plt.grid(True)
plt.show()

# 绘制邻接矩阵热力图
plt.figure(figsize=(10, 10))
plt.imshow(adj_mat[:50, :50], cmap='viridis')  # 展示前50个节点的邻接关系
plt.colorbar()
plt.title('Adjacency Matrix Heatmap (First 50 nodes)')
plt.xlabel('Node ID')
plt.ylabel('Node ID')
plt.show()



```
